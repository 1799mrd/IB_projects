When viewing the html report, we can see that the quality of readings drops towards the end. In order to improve our data in the future, I will cut off the tails of poor quality

Also, amp_res_1.fastq is bad in the quality of per tile sequence

## Filtering the reads with trimmomatic

Cut bases off the start of a read if quality below 20
Cut bases off the end of a read if quality below 20
Trim reads using a sliding window approach, with window size 10 and average quality  within the window 20.
Drop the read if it is below length 20.


```python
!conda install -c bioconda trimmomatic -y
```

```python
!trimmomatic PE -phred33 amp_res_1.fastq amp_res_2.fastq trimmed_20.paired.R1.fastq trimmed_20.unpaired.R1.fastq trimmed_20.paired.R2.fastq trimmed_20.unpaired.R2.fastq.gz ILLUMINACLIP:TruSeq3-PE.fa:2:30:10 LEADING:20 TRAILING:20 SLIDINGWINDOW:10:20 MINLEN:20
```

```python
!trimmomatic PE -phred33 amp_res_1.fastq amp_res_2.fastq trimmed_30.paired.R1.fastq trimmed_30.unpaired.R1.fastq trimmed_30.paired.R2.fastq trimmed_30.unpaired.R2.fastq.gz ILLUMINACLIP:TruSeq3-PE.fa:2:30:10 LEADING:30 TRAILING:30 SLIDINGWINDOW:10:30 MINLEN:30
```
